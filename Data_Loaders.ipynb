## Acreom Loader 

### acreom is a dev-first knowledge base with tasks running on local markdown files.

### Below is an example on how to load a local acreom vault into Langchain. As the local vault in acreom is a folder of plain text .md files, the loader requires the path to the directory.

### Vault files may contain some metadata which is stored as a YAML header. These values will be added to the document’s metadata if collect_metadata is set to true.

```python
from langchain_community.document_loaders import AcreomLoader
```

```python
acreom_loader = AcreomLoader('D:\Data_Science_Study\Course_Pracitse_Code\GenerativeAi\LangChain_Basic\data\Test_Valut',collect_metadata=False)
```

```python
acreom_data = acreom_loader.load()
```

```python
print(acreom_data[0].page_content)
```

## ArxivLoader

### arXiv is an open-access archive for 2 million scholarly articles in the fields of physics, mathematics, computer science, quantitative biology, quantitative finance, statistics, electrical engineering and systems science, and economics.

```python
from langchain_community.document_loaders import ArxivLoader
```

```python
loader = ArxivLoader(query='reasoning',
                     load_max_document = 2,
                     )
```

```python
arxiv_data = loader.load()
```

```python
print(arxiv_data[0].page_content)
```

```python
doc = loader.get_summaries_as_docs()
```

```python
print(doc[0].page_content)
```

## AsyncHtml

### AsyncHtmlLoader loads raw HTML from a list of URLs concurrently.

```python
from langchain_community.document_loaders import AsyncHtmlLoader
```

```python
urls = ["https://www.espn.com", "https://lilianweng.github.io/posts/2023-06-23-agent/"]
```

```python
loader_html = AsyncHtmlLoader(urls)
docs = loader_html.load()
```

```python
print(docs[0].page_content[0:1500])
```

## Pandas DataFrame


```python
from langchain_community.document_loaders import DataFrameLoader
import pandas as pd
```

```python
df = pd.read_csv("data\Street_Tree_List.csv")
```

```python
loader = DataFrameLoader(df,page_content_column='qSpecies')
```

```python
# Use lazy load for larger table, which won't read the full table into memory
for i in loader.lazy_load():
    print(i)
```
